## Problem 3

For the inhomogeneous linear equation.

$$u'(t)=\lambda u(t) + a,\quad a,u(t)\in\mathbb{R}.$$

$$u_n = R(\lambda h)^n u_0 + \left[\frac{R(\lambda h)^n-1}{R(\lambda h)-1}\right] h b^T(I-zA)^{-1}e a.$$

Here $e=(1,\ldots,1)^T\in\mathbb{R}^{r}$,

$$R(z)=1+zb^T(I-zA)^{-1}e.$$

b) Assume $|R(z)|<1$, prove that 

$$\lim_{n\to\infty} u_n=-\frac{a}{\lambda}.$$

Prove that the true solution of the ODE has the same limit, i.e.

$$\lim_{t\to+\infty} u(t)=-\frac{a}{\lambda}.$$

**(a)**

The ODE
$$
f(u,t) = u'(t)=\lambda u(t) + a,\quad a,u(t)\in\mathbb{R} \\
$$

Runga Kutta method  
$$
\begin{align*}
\mathbf{\vec k} &= f(u_n \vec e + hA\vec k) \\ 
      &= \lambda u_n \vec e + \lambda h A \mathbf{ \vec k} + a\vec e \\
u_{n+1} &= u_n + h \vec b^\top \vec k
\end{align*}
$$

Solving for $\vec k$ and letting $z = \lambda h$
$$
\begin{align*}
\vec k &= \lambda (I-zA)^{-1} \vec e u_n + (I-zA)^{-1}a\vec e \\
\implies u_{n+1} &= u_n + h \vec b^\top \left( \lambda (I-zA)^{-1} \vec e u_n + (I-zA)^{-1}a\vec e \right) \\
    &= \left( 1+z\vec b^\top (I-zA)^{-1}\vec e \right)u_n + h\vec b^\top (I-zA)^{-1}a\vec e \\
    &= R(z)u_n + h\vec b^\top (I-zA)^{-1}a\vec e
\end{align*}
$$

Solve for $u_n$ in terms of $u_0$ recursively  
$$
\begin{align*}
u_n &= R(z)\left( R(z)u_{n-1} + h\vec b^\top (I-zA)^{-1}a\vec e \right) + h\vec b^\top (I-zA)^{-1}a\vec e \\
 &= R^2(z)u_{n-1} + \left( R(z) + 1 \right) h\vec b^\top (I-zA)^{-1}a\vec e \\
 &= R^2(z)u_{n-1} + \left( R(z) + 1 \right) \frac{R(z) - 1}{R(z)-1} h\vec b^\top (I-zA)^{-1}a\vec e \\
 &= R^2(z)u_{n-1} + \frac{(R^2(z) - 1)}{R(z)-1}h\vec b^\top (I-zA)^{-1}a\vec e \\
 &= R^2(z)\left( R(z)u_{n-2} + h\vec b^\top (I-zA)^{-1}a\vec e \right) + \frac{(R^2(z) - 1)}{R(z)-1}h\vec b^\top (I-zA)^{-1}a\vec e \\

 &= R^3(z) u_{n-2} + \left( \frac{R^3(z)-R^2(z)}{R(z)-1} + \frac{R^2(z) - 1}{R(z)-1} \right) h\vec b^\top (I-zA)^{-1}a\vec e \\
 &= R^3(z) u_{n-2} +  \frac{R^3(z)-1}{R(z)-1}  h\vec b^\top (I-zA)^{-1}a\vec e \\
 &= \dots \\
 &= R^n(z) u_0  + R^{n-1}(z) \frac{R(z)-1}{R(z)-1} h\vec b^\top (I-zA)^{-1}a\vec e + \frac{(R^{n-1}(z) - 1)}{R(z)-1}h\vec b^\top (I-zA)^{-1}a\vec e \\
 &= R^{n}(z) u_{0} + \left( \frac{R^{n}(z)-R^{n-1}(z)}{R(z)-1} + \frac{R^{n-1}(z) - 1}{R(z)-1} \right) h\vec b^\top (I-zA)^{-1}a\vec e  \\
 &= R^n(z) u_{0} +  \frac{R^n(z)-1}{R(z)-1}  h\vec b^\top (I-zA)^{-1}a\vec e
\end{align*}
$$

**b**  

Runge Kutta Method
$$
\begin{align*}
\lim_{n\to\infty} u_n &= \lim_{n\to\infty} \left( R^n(z) u_{0} +  \frac{R^n(z)-1}{R(z)-1}  h\vec b^\top (I-zA)^{-1}a\vec e \right) \\
&= \lim_{n\to\infty} R^n(z) u_0 + \lim_{n\to\infty} \frac{R^n(z)-1}{R(z)-1}  h\vec b^\top (I-zA)^{-1}a\vec e \\
&= \frac{-1}{R(z)-1} h\vec b^\top (I-zA)^{-1}a\vec e \\
& = \frac{-1}{\lambda h\vec b^\top (I-zA)^{-1}a\vec e} h\vec b^\top (I-zA)^{-1}a\vec e \\
&= -\frac{a}{\lambda}
\end{align*}
$$

True solution
$$
u'(t)=\lambda u(t) + a \\
u'(t) - \lambda u(t) = a \\
$$

Multiplying both sides by $e^{-\lambda t}$
$$
u'(t)e^{-\lambda t} - \lambda u(t)e^{-\lambda t} = a e^{-\lambda t} \\
\left(u(t)e^{-\lambda t}\right)' = a e^{-\lambda t} \\
$$

Integrating
$$
\implies u(t)e^{-\lambda t} = a \int_0^t  e^{-\lambda s} ds \\
u(t) = -\frac{a}{\lambda} \left( 1 + e^{\lambda t} \right)
$$

Taking the limit and remembering that $Re(\lambda)<0$
$$
\begin{align*}
\lim_{t\to\infty} u(t) &= \lim_{t\to\infty} -\frac{a}{\lambda} \left( 1 + e^{\lambda t} \right) \\
&= -\frac{a}{\lambda}
\end{align*}

u(x,0)=e^{-5(x-\pi)^2}, \quad u(0,t)=u(2\pi,t), u_x(0,t)=u_x(2\pi,t)

Use the central finite difference scheme to discretize $u_{xx}(x,t)$, with a uniform grid spacing $\Delta x=2\pi/50$, i.e.

$$
u_{xx}(x_i,t_n)\approx \frac{u(x_i+\Delta x,t_n)-2u(x_i,t_n)+u(x_i-\Delta x,t_n)}{\Delta x^2}.
$$

Implement the following embedded Runge-Kutta method with the Butcher table
$$
\begin{array}{c|ccc}{0} & {} \\ {\frac{2}{3}} & {\frac{2}{3}} \\ {\frac{2}{3}} & {0} & {\frac{2}{3}} \\ \hline & {\frac{1}{4}} & {\frac{3}{4}} \\ \hline  & {\frac{1}{4}}  & {\frac{3}{8}} & {\frac{3}{8}}\end{array}
$$
It has order $p=2$ (propagator), and $\tilde{p}=3$ (error estimator). 


1) Obtain the solution $u(x,T)$ at $T=1.0$ using an adaptive time stepping strategy (with the "Milne device").  The initial time step is $h=0.001$. Plot out the history of the size step along the history of the simulation. Enjoy the simplicity compared to the Trapezoidal-AB2 method.

2) What if you refine the grid spacing to $\Delta x=2\pi/200$? 

**Note**: You should implement this example using **sparse matrices** whenever possible. See 

https://docs.julialang.org/en/v1/stdlib/SparseArrays/index.html

for references on sparse linear algebra.

```julia
using LinearAlgebra
using SparseArrays
using PyPlot

function Laplacian(S)
    L = spzeros(S, S)
    L[diagind(L,0)]  .= 2 # put 2 along diagonal
    L[diagind(L,1)] .= 1 # put 1 along superdiagonal
    L[diagind(L,-1)] .= 1 # put 1 along subdiagonal
    # deal with boundary conditions
    L[1,end] = 1
    L[end,1] = 1
    # L
    return L
end

function RkAdapt(N,S,T,u0)
    u = spzeros(S, N)
    u[:, 1] = u0
    h = T/N

    b1 = [1/4, 3/4, 0]
    b2 = [1/4, 3/8, 3/8]
    A = [0 0 0; 2/3 0 0; 0 2/3 0]
    L = Laplacian(S)/((2*pi/S)^2)
    H = [h]
    tol = 10^(-9)
    y = spzeros(S)
    for i = 1:N-1
        # milne device
        if 1 in (y-u[:,i] .> tol) # if error is too big decrease step size
            h = h/2
            append!(H,h)
            if i != 1
                i = i-1
            end
        elseif 1 in (y-u[:,i] .< 1/10 * tol) # if error is too small increase step size
            h = h*2
            append!(H,h)
        end
        k1 = L*u[:,i]
        k2 = L*u[:,i] + 2*h/3 *k1
        k3 = L*u[:,i] + 3*h/3 *k2
#         k = [k1,k2,k3]
        
        u[:,i+1] = u[:,i] + h*(1/4*k1 + 3/8*k2)
        y = u[:,i] + h*(1/4*k1 + 3/8*k2 + 3/8*k3)
    end
    return u,H
end

N = 1000 # number of steps
S = 100 # number of spaces in grid
T = 1
a = 0
b = 2*pi
f(x) = exp(-5*(x-pi)^2)
xs = range(a,stop=b,length=S)
u₀ = [f(i) for i in xs] # discritized initial condition
A,H = RkAdapt(N,S,T,u₀)

figure(1,figsize=(5,5))
plot(xs[1:99], A[1:99, 2],"r-",label=L"ans")
H
```
