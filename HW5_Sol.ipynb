## Problem 1

For a separable Hamiltonian system,

$$H(p,q) = T(p)+V(q),\quad p,q\in \mathbb{R}^d$$

with corresponding Hamiltonian equation

$$p'(t)=-\nabla_q V(q),\quad q'(t)=\nabla_p T(p),$$


prove that the symplectic Euler method 

$$
\begin{split}
p_{n+1} &= p_n - \Delta t \nabla_q V(q_n),\\
q_{n+1} &= q_n + \Delta t \nabla_p T(p_{n+1}),
\end{split}
$$

is 

a) an order 1 method.

b) a symplectic method.

Hint: 

Try to prove
$$
\Phi_{n+1}^{\top} J \Phi_{n+1}=\Phi_{n}^{\top} J \Phi_{n}.
$$

A useful fact is that for any symmetric matrix $X\in \mathbb{R}^{d\times d}$,

$$  \begin{pmatrix}
I & 0\\
X & I
\end{pmatrix}^T J  \begin{pmatrix}
I & 0\\
X & I
\end{pmatrix} = J, \quad  \begin{pmatrix}
I & X\\
0 & I
\end{pmatrix}^T J  \begin{pmatrix}
I & X\\
0 & I
\end{pmatrix} = J.
\tag{SymFact}
$$

**Proof**:

a) Check the local truncation error. Taylor expansion and find

$$
p(t_{n+1})-p(t_n)+\Delta t \nabla_q V(q(t_n)) = O(\Delta t^2).
$$

$$
q(t_{n+1}) - q(t_n) - \Delta t \nabla_p T(p(t_{n+1})) = O(\Delta t^2).
$$

where $p(t),q(t)$ are the true solution to the Hamiltonian system.

b) Define 

$$\Phi_n = 
\begin{pmatrix}
\frac{\partial p_n}{\partial p_0} & \frac{\partial p_n}{\partial q_0}\\
\frac{\partial q_n}{\partial p_0} & \frac{\partial q_n}{\partial q_0}
\end{pmatrix}.
$$

The symplectic Euler scheme can be rewritten as

$$
\begin{split}
p_{n+1} &= p_n - \Delta t \nabla_q V(q_n),\\
-\Delta t \nabla_p T(p_{n+1})+q_{n+1} &= q_n.
\end{split}
$$

Hence

$$
\begin{pmatrix}
I & 0\\
-\Delta t \nabla_{pp} T(p_{n+1}) & I
\end{pmatrix} \Phi_{n+1} 
= \begin{pmatrix}
I & - \Delta t \nabla_{qq} V(q_n)\\
0 & I
\end{pmatrix}\Phi_n.
$$

or 

$$
\Phi_{n+1} 
= \begin{pmatrix}
I & 0\\
+\Delta t \nabla_{pp} T(p_{n+1}) & I
\end{pmatrix} 
\begin{pmatrix}
I & - \Delta t \nabla_{qq} V(q_n)\\
0 & I
\end{pmatrix}\Phi_n.
$$

Then use (SymFact), we have 

$$
\Phi_{n+1}^{\top} J \Phi_{n+1}=\Phi_{n}^{\top} J \Phi_{n}.
$$

## Problem 2

Prove that the 1st stage Gauss-Legendre method is symplectic.

Hint:

The Gauss-Legendre method is for Hamiltonian system is
$$
\begin{align*}k_{n}=&f\left(u_{n}+\frac{1}{2} h k_{n}\right)=J^{-1} \nabla H\left(u_{n}+\frac{1}{2} h k_{n}\right),\\u_{n+1}=&u_{n}+h k_{n}\end{align*}
$$
You should consider both derivatives 


$$
\Phi_{n}=\frac{\partial u_{n}}{\partial u_{0}}, \quad \Xi_{n}=\frac{\partial k_{n}}{\partial u_{0}}.
$$

You can also assume that 
$$
G_n= \nabla^{2} H\left(u_{n}+\frac{h}{2} k_{n}\right)
$$
is invertible. 
Then try to prove
$$
\Phi_{n+1}^{\top} J \Phi_{n+1}=\Phi_{n}^{\top} J \Phi_{n}.
$$

**Proof**: 

Taking the derivatives
$$
\begin{align*}
\Xi_{n}=&J^{-1} G_n\left(\Phi_{n}+\frac{1}{2} h \Xi_{n}\right),\\
\Phi_{n+1}=&\Phi_{n}+h \Xi_{n}.
\end{align*}
$$
Then
$$
\Phi_{n}=G_{n}^{-1} J \Xi_{n}-\frac{1}{2} h \Xi_{n}.
$$
Then
$$
\begin{align*}
\Phi_{n+1}^{\top} J \Phi_{n}=&\left(\Phi_{n}^{\top}+h \Xi_{n}^{\top}\right) J\left(\Phi_{n}+h \Xi_{n}\right)\\
=&\Phi_n^T J\Phi_n+h\left(\Xi_n^T(-J) G_n^{-1}-\frac12 h\Xi_n^{\top}\right) J\Xi_n\\
&+h\Xi_n^{\top}J(G_n^{-1}J\Xi_n -\frac12 h\Xi_n) + h^2\Xi_n^{\top} J\Xi_n\\
=&\Phi_n^T J\Phi_n.
\end{align*}
$$
Hence 
$$
\Phi_{n+1}^{\top} J \Phi_{n}=\Phi_n^T J\Phi_n=\cdots=\Phi_0^T J\Phi_0=I.
$$

## Problem 3

Derive the Butcher table for the Gauss-Legendre-2 step method, starting from the roots of the Legendre polynomial $P_2(x)$. 

**Solution**: 

The Legendre polynomial $P_2(x) = \frac{1}{2}(3x^2-1), -1\leq x \leq 1$. The roots of $P_2(x)$ are $\pm\frac{1}{\sqrt{3}}$, meaning the Gaussian quadrature points for $[-1,1]$ are $\pm\frac{1}{\sqrt{3}}$. By changing interval $[-1,1]$ to $[t,t+h]$, the quadrature points for $[t,t+h]$ become $t + (\frac{1}{2}-\frac{\sqrt{3}}{6})h, t + (\frac{1}{2}+\frac{\sqrt{3}}{6})h$. Therefore 
$$c_1 = \frac{1}{2}-\frac{\sqrt{3}}{6}, \quad c_2 = \frac{1}{2}+\frac{\sqrt{3}}{6}.$$

The corresponding Lagrange interpolation polynomials are 
$$l_1(s) = \frac{s-(t+c_2h)}{(c_1-c_2)h}, \quad l_2(s) = \frac{s-(t+c_1h)}{(c_2-c_1)h}.$$ 
Then the cofficients of the Gauss-Legendre-2 step method are
$$A_{11} = \frac{1}{h} \int_t^{t+c_1h}l_1(s)ds = \frac{c_1(c_1-2c_2)}{2(c_1-c_2)} = \frac{1}{4},$$
$$A_{12} = \frac{1}{h} \int_t^{t+c_1h}l_2(s)ds = -\frac{c_1^2}{2(c_2-c_1)} = \frac{1}{4}-\frac{\sqrt{3}}{6},$$
$$A_{21} = \frac{1}{h} \int_t^{t+c_2h}l_1(s)ds = \frac{c_2^2}{2(c_2-c_1)} = \frac{1}{4}+\frac{\sqrt{3}}{6},$$
$$A_{22} = \frac{1}{h} \int_t^{t+c_2h}l_2(s)ds = \frac{c_2(c_2-2c_1)}{2(c_2-c_1)} = \frac{1}{4},$$
$$b_1 = \frac{1}{h}\int_t^{t+h}l_1(s)ds = \frac{1-2c_2}{2(c_1-c_2)} = \frac{1}{2},$$
$$b_2 = \frac{1}{h}\int_t^{t+h}l_2(s)ds = \frac{1-2c_1}{2(c_2-c_1)} = \frac{1}{2}.$$

## Problem 4


Consider two atoms with Lennard-Jones interaction

$$
V(x_1,x_2)=w(|x_1-x_2|),\quad w(r)=0.01\left(\frac{1}{r^{12}}-\frac{1}{r^6}\right).
$$

a) Use the 1-step (2nd order) Gauss-Legendre method to solve the Hamiltonian system

$$
x_1''(t) = -\nabla_{x_1} V(x_1,x_2), \quad x_2''(t) = -\nabla_{x_2} V(x_1,x_2),
$$

with initial condition

$$
x_1(0) = 0.7,\quad x_2(0)=-0.7, \quad \dot{x}_1(0)=\dot{x}_2(0)=0.0.
$$

Use time step $h=0.2$ on the time interval $[0,50.0]$. Plot the distance $x_1(t)-x_2(t)$ along the trajectory, and plot the energy along the trajectory
$$
E(t)=\frac12 (x_1'(t))^2 + \frac12 (x_2'(t))^ 2 + V(x_1(t),x_2(t)).
$$

You can use any solver to solve the nonlinear equation (but you have to code from scratch)

b) Do the same calculation using the Velocity-Verlet method, (and observe the ease of implementation compared to implicit methods).



```julia
using LinearAlgebra

function force(x)
    r = abs(x[1]-x[2])
    f1 = +0.01*(12.0./r.^13 - 6.0./r.^7).*(x[1]-x[2])./r
    f2 = -f1
    return [f1,f2]
end

function potential(x)
    r = abs(x[1]-x[2])
    return  +0.01*(1.0/r.^12 - 1.0/r.^6)
end
    
function gl1(N,T,x0,v0,Niter)
    dt = T/N
    x = zeros(2,N+1)
    v = zeros(2,N+1)
    f = zeros(2,N+1)
    Vpot = zeros(N+1)
    
    t = collect(0:N)*dt
    x[:,1] = x0
    v[:,1] = v0
    f[:,1] = force(x[:,1])
    Vpot[1] = potential(x[:,1])
    
    tol = 1e-10
    for i = 1 : N
        # Use fixed point iteration to solve the nonlinear equation
        # w = u_n + 0.5 * dt * k
        wx = x[:,i]
        wv = v[:,i]
        l=1
        res = 0.0
        for l = 1 : Niter
            wxNew = x[:,i] + 0.5*dt*wv
            wvNew = v[:,i] + 0.5*dt*force(wx)
            res = sqrt(norm(wxNew-wx)^2 + norm(wvNew-wv)^2)
            if( res < 1e-10 )
                break
            end
            wx = wxNew
            wv = wvNew
        end
        kx = (wx - x[:,i]) / (0.5*dt)
        kv = (wv - v[:,i]) / (0.5*dt)
        x[:,i+1] = x[:,i] + dt*kx
        v[:,i+1] = v[:,i] + dt*kv
        Vpot[i+1] = potential(x[:,i+1])
    end
    return x,v,Vpot
end

T = 50.0
dt = 0.2
N = round(Int64,T/dt)
t = collect(0:N)*dt
x0 = [0.7,-0.7]
v0 = [0.0,0.0]

x,v,Vpot = gl1(N,T,x0,v0,100)

using PyPlot
figure(1,figsize=(5,5))
st=1
plot(t,vec(x[1,:]-x[2,:]),"bo")
xlabel(L"t")
ylabel(L"x_1-x_2")
title("GL1")

Et = zeros(N+1)
Et = 0.5*vec(v[1,:].^2+v[2,:].^2) + Vpot

figure(2,figsize=(5,5))
st=1
plot(t,Et,"b-")
xlabel(L"t")
ylabel(L"E(t)")
title("GL1")
```

```julia
# Verlet
function verlet(N,T,x0,v0)
    dt = T/N
    x = zeros(2,N+1)
    v = zeros(2,N+1)
    f = zeros(2,N+1)
    Vpot = zeros(N+1)
    t = collect(0:N)*dt
    x[:,1] = x0
    v[:,1] = v0
    f[:,1] = force(x[:,1])
    Vpot[1] = potential(x[:,1])
    for i = 1 : N
        v[:,i+1] = v[:,i] + 0.5*dt*f[:,i]
        x[:,i+1] = x[:,i] + dt*v[:,i+1]
        f[:,i+1] = force(x[:,i+1])
        Vpot[i+1] = potential(x[:,i+1])
        v[:,i+1] = v[:,i+1] + 0.5*dt*f[:,i+1]
    end
    return x,v,Vpot
end

T = 50.0
dt = 0.2
N = round(Int64,T/dt)
t = collect(0:N)*dt
x0 = [0.7,-0.7]
v0 = [0.0,0.0]

x,v,Vpot = verlet(N,T,x0,v0)

Et = zeros(N+1)
Et = 0.5*vec(v[1,:].^2+v[2,:].^2) + Vpot

using PyPlot
figure(1,figsize=(5,5))
st=1
plot(t,vec(x[1,:]-x[2,:]),"b-o")
xlabel(L"t")
ylabel(L"x_1-x_2")
title("Verlet")

figure(2,figsize=(5,5))
st=1
plot(t,Et,"b-")
xlabel(L"t")
ylabel(L"E(t)")
title("Verlet")
```
