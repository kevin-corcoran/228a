## Problem 1

For a separable Hamiltonian system,

$$H(p,q) = T(p)+V(q),\quad p,q\in \mathbb{R}^d$$

with corresponding Hamiltonian equation

$$p'(t)=-\nabla_q V(q),\quad q'(t)=\nabla_p T(p),$$


prove that the symplectic Euler method 

$$
\begin{split}
p_{n+1} &= p_n - \Delta t \nabla_q V(q_n),\\
q_{n+1} &= q_n + \Delta t \nabla_p T(p_{n+1}),
\end{split}
$$

is 

a) an order 1 method.

b) a symplectic method.

Hint: 

Try to prove
$$
\Phi_{n+1}^{\top} J \Phi_{n+1}=\Phi_{n}^{\top} J \Phi_{n}.
$$

A useful fact is that for any symmetric matrix $X\in \mathbb{R}^{d\times d}$,

$$  \begin{pmatrix}
I & 0\\
X & I
\end{pmatrix}^T J  \begin{pmatrix}
I & 0\\
X & I
\end{pmatrix} = J, \quad  \begin{pmatrix}
I & X\\
0 & I
\end{pmatrix}^T J  \begin{pmatrix}
I & X\\
0 & I
\end{pmatrix} = J.
\tag{SymFact}
$$

Flow map $\phi_n$  

$$
\begin{align*}
\phi_n &\dot= \begin{pmatrix} p_{n+1} \\ q_{n+1} \end{pmatrix} 
              = \begin{pmatrix} p_{n} - h\nabla_q V(q_n) \\ q_{n} + h\nabla_p T(p_{n+1}) \end{pmatrix} \\
\implies \nabla\phi_n &= \begin{pmatrix} \nabla p_{n+1} \\ \nabla q_{n+1} \end{pmatrix}

## Problem 4



Consider two atoms with Lennard-Jones interaction

$$
V(x_1,x_2)=w(|x_1-x_2|),\quad w(r)=0.01\left(\frac{1}{r^{12}}-\frac{1}{r^6}\right).
$$

a) Use the 1-step (2nd order) Gauss-Legendre method to solve the Hamiltonian system

$$
x_1''(t) = -\nabla_{x_1} V(x_1,x_2), \quad x_2''(t) = -\nabla_{x_2} V(x_1,x_2),
$$

with initial condition

$$
x_1(0) = 0.7,\quad x_2(0)=-0.7, \quad \dot{x}_1(0)=\dot{x}_2(0)=0.0.
$$

Use time step $h=0.2$ on the time interval $[0,50.0]$. Plot the distance $x_1(t)-x_2(t)$ along the trajectory, and plot the energy along the trajectory
$$
E(t)=\frac12 (x_1'(t))^2 + \frac12 (x_2'(t))^ 2 + V(x_1(t),x_2(t)).
$$

You can use any solver to solve the nonlinear equation (but you have to code from scratch)

b) Do the same calculation using the Velocity-Verlet method, (and observe the ease of implementation compared to implicit methods).

```julia
using LinearAlgebra

function newton(F, k, tol, maxIter)
    # F = 0 fixed point, # k v
    for i = 1:maxIter
        x = k
        k = F(x)
        condition = abs.(x - k) .> tol
        if (condition[1] == 0 && condition[2] == 0)
            break
        end
    end
    return k
end

function V(x)
    r = abs(x[1]-x[2])
    return 0.01*(1/r^12 - 1/r^6)
end

function gradV(x, v)
    r = abs(x[1]-x[2]) # do we need absolute values
    x[1] = v[1]
    x[2] = v[2]
    v[1] = -0.01*(-12*v[1]/(r^13) + 6*v[1]/(r^7))
    v[2] = -0.01*(12*v[2]/(r^13) - 6*v[2]/(r^7))
    return x, v
end

function gl1(N, T, x0, v0, maxIter)
    h = T/N
    
    x = zeros(2,N+1)
    v = zeros(2,N+1)
    Vpot = zeros(1,N+1)
    x[:,1] = x0
    v[:,1] = v0
    Vpot[1] = V(x0)
    Niter = zeros(N)
    
    kv = zeros(2, N+1)
    kv[:, 1] = [0.1, 0.1] # initial k value guess
    tol = [10^(-16), 10^(-16)]
        
    A = [0.5]
    b = [1]
    for i = 1:N
#         x, v = gradV(x[:,i]+h/2*k, v[:,i]+h/2*k)
        F(k) = k - gradV(x[:,i]+h/2*k, v[:,i]+h/2*k)[2]
        kv[:, i+1] = newton(F, kv[:, i], tol, maxIter)
        v[:,i+1] = v[:,i] + h*kv[:,i+1]
        x[:, i+1] = x[:,i] + h*v[:,i]
        Vpot[i+1] = V(x[:,i+1])
    end
    return x, v, Vpot
end

T = 50.0
dt = 0.2
N = round(Int64,T/dt)
t = collect(0:N)*dt
x0 = [0.7,-0.7]
v0 = [0.0,0.0]

x,v,Vpot = gl1(N,T,x0,v0,100)

using PyPlot
figure(1,figsize=(5,5))
st=1
plot(t,vec(x[1,:]-x[2,:]),"bo")
xlabel(L"t")
ylabel(L"x_1-x_2")
title("GL1")

Et = zeros(N+1)
Et = 0.5*vec(v[1,:].^2+v[2,:].^2) + Vpot'

figure(2,figsize=(5,5))
st=1
plot(t,Et,"b-")
xlabel(L"t")
ylabel(L"E(t)")
title("GL1")
```

**part (b)**  
Velocity-Verlet

```julia
function verlet(N, T, x0, v0)
    h = T/N
    
    x = zeros(2,N+1)
    v = zeros(2,N+1)
    Vpot = zeros(1,N+1)
    x[:,1] = x0
    v[:,1] = v0
    Vpot[1] = V(x0)
    
    for i = 1:N
        xt, vt = gradV(x[:,i], v[:,i])
        vhalf = v[:,i]-h/2 *xt
        x[:, i+1] = x[:,i] + h *vhalf
        v[:,i+1] = v[:,i] - h/2 *x[:,i+1]
        Vpot[i+1] = V(x[:,i+1])
    end
    return x, v, Vpot
end

T = 50.0
dt = 0.2
N = round(Int64,T/dt)
t = collect(0:N)*dt
x0 = [0.7,-0.7]
v0 = [0.0,0.0]

x,v,Vpot = verlet(N,T,x0,v0)

figure(1,figsize=(5,5))
st=1
plot(t,vec(x[1,:]-x[2,:]),"bo")
xlabel(L"t")
ylabel(L"x_1-x_2")
title("verlet")

Et = zeros(N+1)
Et = 0.5*vec(v[1,:].^2+v[2,:].^2) + Vpot'

figure(2,figsize=(5,5))
st=1
plot(t,Et,"b-")
xlabel(L"t")
ylabel(L"E(t)")
title("Verlet")
```
