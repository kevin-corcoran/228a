## Problem 1

The 2-stage Milne's method is

$$u_{n+1}=u_{n-1}+h\left(\frac{1}{3}f_{n+1}+\frac{4}{3}f_n+\frac{1}{3}f_{n-1}\right).$$

Prove with analytic computation that 

1. It is a 4th order method.

2. The region of absolute stability of is only along the imaginary axis.

**Solution:**

Taylor expansion 

$$z^2-1-\log z (\frac{1}{3}z^2+\frac{4}{3}z+\frac{1}{3})=-\frac{(z-1)^5}{90}+O((z-1)^6).$$

Use the boundary locus method, and the boundary is

$$
\begin{split}
z&=\frac{e^{2i\theta}-1}{\frac13e^{2i\theta}+\frac43e^{i\theta}+\frac13}\\
&=\frac{3i\sin\theta}{\cos\theta+2}.
\end{split}
$$

So the boundary is along the imaginary axis. Plug in $z=\pm 1$ to show that neither side of the imaginary axis is in the stability region.

Note: if one chooses $z=3$, then the $u_{n+1}$ is eliminated from the equation, and $u_{n+1}$ can choose any value. Therefore $z=3$ does not belong to the RAS.

## Problem 3

For the inhomogeneous linear equation.

$$u'(t)=\lambda u(t) + a,\quad a,u(t)\in\mathbb{R}.$$

Consider the Runge-Kutta method with Butcher tableau $A\in\mathbb{R}^{r\times r},b\in\mathbb{R}^{r},Re\lambda<0$

a) Prove that the numerical solution can be written as

$$u_n = R(\lambda h)^n u_0 + \left[\frac{R(\lambda h)^n-1}{R(\lambda h)-1}\right] h b^T(I-zA)^{-1}e a.$$

Here $e=(1,\ldots,1)^T\in\mathbb{R}^{r}$,

$$R(z)=1+zb^T(I-zA)^{-1}e.$$

b) Assume $|R(z)|<1$, prove that 

$$\lim_{n\to\infty} u_n=-\frac{a}{\lambda}.$$

Prove that the true solution of the ODE has the same limit, i.e.

$$\lim_{t\to+\infty} u(t)=-\frac{a}{\lambda}.$$

**Proof:**

### a)

We know the Runge-Kutta method is as follows:

\begin{eqnarray}
k_1 &=& \lambda (u_n + h(a_{11} k_1 + a_{12} k_2 + \cdots + a_{1r} k_r) + a\\
k_2 &=& \lambda (u_n + h(a_{21} k_1 + a_{22} k_2 + \cdots + a_{2r} k_r) + a\\
\cdots &=& \cdots \\
k_r &=& \lambda (u_n + h(a_{r1} k_1 + a_{r2} k_2 + \cdots + a_{rr} k_r) + a\\
u_{n+1} &=& u_n + h(b_1 k_1 + b_2k_2 + \cdots + b_rk_r)
\end{eqnarray}

Denote $e=(1,\cdots,1)^T\in\mathbb{R}^{r}$ and rewrite Runge-Kutta method as a matrix form: 
\begin{eqnarray}
\mathbf k  &=& \lambda u_n e + \lambda h A \mathbf k + a e\\
u_{n+1} &=& u_n + h b^T \mathbf k
\end{eqnarray}
Solve the first equation, we can get $\mathbf k = (I-\lambda h A)^{-1} \lambda u_n e + (I-\lambda h A)^{-1} ae.$ Thus, substitute $\mathbf k$ into the second equation, we can get 

\begin{align*}
u_{n+1}& = u_n + hb^T ((I-\lambda h A)^{-1} \lambda u_n e) + hb^T (I-\lambda h A)^{-1}ae \\
&= (1+ \lambda h b^T (I-\lambda h A)^{-1} e )u_n + hb^T  (I-\lambda h A)^{-1}ae \\
&= R(\lambda h) u_n + hb^T (I-\lambda h A)^{-1}ea 
\end{align*}
Apply this equation recursively, we can get 
\begin{align*}
u_{n}&= R(\lambda h) u_{n-1} + hb^T (I-\lambda h A)^{-1}ea \\
&= R(\lambda h)^2 u_{n-2} + (1+ R(\lambda h) )hb^T (I-\lambda h A)^{-1}ea \\
&= R(\lambda h)^3 u_{n-3} + (1+ R(\lambda h)+ R(\lambda h)^2 )hb^T (I-\lambda h A)^{-1}ea \\
& \cdots \\
&= R(\lambda h)^n u_{0} + (1+ R(\lambda h)+ R(\lambda h)^2 + \cdots + R(\lambda h)^{n-1} )hb^T (I-\lambda h A)^{-1}ea \\
&= R(\lambda h)^n u_{0} + \left[\frac{R(\lambda h)^n-1}{R(\lambda h)-1}\right]hb^T (I-\lambda h A)^{-1}ea. \\
\end{align*}

Therefore, the result of problem a) holds.


### b)

Since $|R(z)|<1$ and $|R(z)^n|\le |R(z)|^n$, we know that the limit $\lim_{n\to \infty}  R(z)^n =0$.

We collect the term for $R(\lambda h)^n$, it is $v_n=R(\lambda h)^n u_{0} + \left[\frac{R(\lambda h)^n}{R(\lambda h)-1}\right]hb^T (I-\lambda h A)^{-1}ea,$ which vanishes as $n\to \infty$.

Considering the residual in $u_n$, 
\begin{align*}
w_n &= \left[\frac{1}{-R(\lambda h)+1}\right]hb^T (I-\lambda h A)^{-1}ea.\\
&= \left[\frac{1}{-(1+\lambda hb^T(I-\lambda hA)^{-1}e)+1}\right]hb^T (I-\lambda h A)^{-1}ea.\\
&= \left[\frac{1}{-\lambda hb^T(I-\lambda hA)^{-1}e}\right]hb^T (I-\lambda h A)^{-1}ea.\\
&= -\frac{a}{\lambda}.
\end{align*}
Therefore, $\lim_{n\to \infty} u_n = \lim_{n\to \infty} v_n + \lim_{n\to \infty} w_n = 0 + (-\frac{a}{\lambda} )= -\frac{a}{\lambda}.$

Now, we consider the true solution of differential equation. 




\begin{eqnarray}
&u'(t)&=&\lambda u(t) + a\\
&u'(t) - \lambda u(t)&=&a\\
\end{eqnarray}
Multiply both sides with $e^{-\lambda t}$, we can get 
\begin{eqnarray}
&e^{-\lambda t} u'(t) - \lambda e^{-\lambda t} u(t)&=&ae^{-\lambda t}\\
&(e^{-\lambda t} u(t))'&=&ae^{-\lambda t}\\
\end{eqnarray}
Integrate both sides from 0 to t, we can get 
\begin{eqnarray}
&\int_0^t (e^{-\lambda t} u(t))' \mathbf d t &=&\int_0^t  ae^{-\lambda t} \mathbf d t \\
& e^{-\lambda t} u(t) - u(0) & = &\int_0^t  ae^{-\lambda t} \mathbf d t \\
& e^{-\lambda t} u(t) - u(0) & = & \frac{a}{-\lambda}( e^{-\lambda t} -1) \\
& e^{-\lambda t} u(t) & = &   u(0) +  \frac{a}{-\lambda}( e^{-\lambda t} -1) \\
&  u(t) & = &   e^{\lambda t} u(0) +  \frac{a}{-\lambda}( 1 - e^{\lambda t} ) \\
\end{eqnarray}
We know that $Re\lambda<0$, thus $\lim_{t\to \infty} e^{\lambda t} = 0.$
Therefore, we can see that $\lim_{t\to \infty} u(t) = -\frac{a}{\lambda}.$

To sum up, the limit of Runge-Kutta scheme is indeed the limit of the true solution.

## Problem 4

Consider the heat equation in 1D with periodic boundary conditions

$$
u_t(x,t) = u_{xx}(x,t),\quad x\in [0,2\pi],
$$

$$
u(x,0)=e^{-5(x-\pi)^2}, \quad u(0,t)=u(2\pi,t), u_x(0,t)=u_x(2\pi,t)
$$

Use the central finite difference scheme to discretize $u_{xx}(x,t)$, with a uniform grid spacing $\Delta x=2\pi/50$, i.e.

$$
u_{xx}(x_i,t_n)\approx \frac{u(x_i+\Delta x,t_n)-2u(x_i,t_n)+u(x_i-\Delta x,t_n)}{\Delta x^2}.
$$

Implement the following embedded Runge-Kutta method with the Butcher table
$$
\begin{array}{c|ccc}{0} & {} \\ {\frac{2}{3}} & {\frac{2}{3}} \\ {\frac{2}{3}} & {0} & {\frac{2}{3}} \\ \hline & {\frac{1}{4}} & {\frac{3}{4}} \\ \hline  & {\frac{1}{4}}  & {\frac{3}{8}} & {\frac{3}{8}}\end{array}
$$
It has order $p=2$ (propagator), and $\tilde{p}=3$ (error estimator). 


1) Obtain the solution $u(x,T)$ at $T=1.0$ using an adaptive time stepping strategy (with the "Milne device").  The initial time step is $h=0.001$. Plot out the history of the size step along the history of the simulation. Enjoy the simplicity compared to the Trapezoidal-AB2 method.

2) What if you refine the grid spacing to $\Delta x=2\pi/200$? 

**Note**: You should implement this example using **sparse matrices** whenever possible. See 

https://docs.julialang.org/en/v1/stdlib/SparseArrays/index.html

for references on sparse linear algebra.

```julia
using LinearAlgebra
using SparseArrays

function RKStep(u,A,dt)
    # Implement one step of the embedded Runge-Kutta method
    k1 = A*u
    k2 = A*(u+dt*2.0/3.0*k1)
    k3 = A*(u+dt*2.0/3.0*k2)
    u_new = u+dt*(1.0/4.0 * k1 + 3.0/4.0 * k2)
    y_new = u+dt*(1.0/4.0 * k1 + 3.0/8.0 * k2 + 3.0 / 8.0 * k3)
    return u_new, y_new
end

function EmbedRK(T,dt0,u0,A,delta)
    dt = T/N
    Ns = length(u0)
    I_Ns = sparse(1.0I,Ns,Ns)
    @assert(Ns == size(A,1) && Ns == size(A,2))
    u = copy(u0)
    
    t = 0.0;
    tHist = [t];
    dt = dt0


    while(t < T)

        if( t + dt >= T )
            # propagate and stop
            dt = T-t
            append!(tHist,T)
            u, _ = RKStep(u,A,dt)
            break
        end
        
        # propagate and compute error
        u_new, y_new = RKStep(u,A,dt)
        kappa = norm(u_new - y_new)

        if( kappa <= 0.6 * dt * delta )
            # Advance
            t = t + dt
            append!(tHist,t)
            dt *= 2.0
        elseif( kappa > 6 * dt * delta)
            # halve the step size. Do not advance
            dt *= 0.5
        else
            # Advance            
            t = t + dt
            append!(tHist,t)
            u = copy(u_new)
        end
        if( dt < 1e-5 )
            break
        end
    end

    return u,tHist
end

Ls = 2*pi
Ns = 50
hs = Ls / Ns
xs = collect(0:Ns-1)*hs

A = 1.0/hs^2 * spdiagm(-1=>ones(Ns-1),0=>-2.0*ones(Ns),1=>ones(Ns-1))
A[1,Ns] = 1.0 / hs^2
A[Ns,1] = 1.0 / hs^2
u0 = exp.(-5.0*(xs .- pi).^2)

T  = 1.0
dt0 = 0.001
delta = 1e-3

uT,tHist = EmbedRK(T,dt0,u0,A,delta)

using PyPlot
figure(1,figsize=(5,5))
plot(xs,vec(u0),"b--",label=L"u(0)")
plot(xs,vec(uT),"r-",label=L"u(T)")
legend()

figure(2,figsize=(5,5))
plot(tHist[1:end-1],diff(tHist),"b-o")
xlabel("t")
ylabel("h")
```
