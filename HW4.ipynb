## Problem 1

Consider the ODE 

$$u'(t) = \lambda (u − \cos t) − \sin t, \quad u(0) = u_0,$$

with the exact solution

$$u(t) = e^{\lambda t}(u_0 − 1) + \cos t.$$

Solve the equation and compare with the exact solution using 

a) The trapezoidal method

b) The backward Euler method

for $0\le t \le 3$. Use the parameter $\lambda=-10^6,u_0=1.5,\Delta t = 0.1$. 

Why trapezoidal rule seems to be less accurate? Repeat with $\Delta t=0.001$. Explain the difference in terms of the behavior $R(z)$ at $z=\infty$.

**Begin**  


Backward Euler can be rewritten as   

$$
\begin{align*}
u_{n+1} &=u_n +hf(u_{n+1},t_{n+1})\\
    &= \frac{1}{1-h\lambda}\left( u_n-h\left( \lambda cos(t_{n+1} +sin(t_{n+1} \right) \right)
\end{align*}
$$

Trapezoidal Rule can be rewritten as  

$$
\begin{align*}
u_{n+1} &=u_n +\frac{h}{2}\left(f(u_n,t_n) + f(u_{n+1},t_{n+1})\right)\\
u_{n+1} &= u_n +\frac{h}{2} \left( \lambda u_n − 2\lambda \cos t_{n+1} − 2\sin t_{n+1} +
                            \lambda u_{n+1} \right) \\
    &=\frac{1}{1-\frac{h\lambda}{2}}\left( \left(1+\frac{h\lambda}{2}\right)u_n
           -h(\lambda cos(t_{n+1} +sin(t_{n+1}) \right)
\end{align*}
$$

```julia
function BackEuler(N,T,u0,lambda)
    u = zeros(1,N+1)
    u[1] = u0
    h = N/T
    λ = lambda
    for i = 1:N-1
        u[i+1] = 1/(1-h*λ)*(u[i] - h*(λ*cos((i)*h)+sin((i)*h)))
    end
    return u
end

function Trape(N,T,u0,lambda)
    u = zeros(1,N+1)
    u[1] = u0
    h = N/T
    λ = lambda
    for i = 1:N-1
        u[i+1] = 1/(1-h*λ/2)*((1+h*λ/2)*u[i] - h*(λ*cos((i)*h)+sin((i)*h)))
    end

    return u
end


u0 = 1.5
lambda = -1.0e+6
T = 3.0
dt = 0.1
N = round(Int64,T/dt)
t = collect(0:N)*dt
uexact = exp.(lambda*t)*(u0-1)+cos.(t)
uBackEuler = BackEuler(N,T,u0,lambda)[:]
uTrape = Trape(N,T,u0,lambda)[:]

using PyPlot
figure(1,figsize=(5,5))
plot(t,uTrape,"r-d",label="Trape")
plot(t,uexact,"k-<",label="exact")
plot(t,uBackEuler,"b-o",label="BackEuler")
legend()

dt = 0.001
N = round(Int64,T/dt)
t = collect(0:N)*dt
uexact = exp.(lambda*t)*(u0-1)+cos.(t)
uBackEuler = BackEuler(N,T,u0,lambda)[:]
uTrape     = Trape(N,T,u0,lambda)[:]
figure(2,figsize=(5,5))
plot(t,uTrape,"r-",label="Trape")
plot(t,uexact,"k-",label="exact")
plot(t,uBackEuler,"b-",label="BackEuler")
legend()
```

$$
m'(x) = 2\left<f(x,u)-f(x,v), u(x)-v(x)\right> \leq 2Lm(x)

Since Gauss-Legendre is derived from a collocation method, let $x_i = x_0 + h c_i$ be the collacation points. Then

$$
m'(x) = 2\left<f(x_i,u(x_i))-f(x_i,v(x_i)), u(x_i)-v(x_i)\right> \leq 0 \tag{1}
$$

Consider

$$

```julia
# Gauss-Legendre 2-step. Fixed point formulation

function fixpoint(k, u, A, eps)
    tol = [10^(-15), 10^(-15)]
    c = 0
    
    g(k) = [u[2] + A[2,1]*k[1] + A[2,2]*k[2], 1/eps *(1-(u[1]+A[1,1]*k[1]+A[1,2]*k[2])^2) * (u[2] + A[2,1]*k[1] + A[2,2]*k[2]) - u[1]+A[1,1]*k[1]+A[1,2]*k[2]]
    while (c < 500)
        x = k
        k = g(x)
        c += 1
        condition = abs.(x - k) .> tol
        if (condition[1] == 0 && condition[2] == 0)
            break
        end
    end
    return k, c
end
    
function gl2fixpoint(N,T,u0,eps,maxIter)
    h = T/N
    u = zeros(2,N+1)
    u[:,1] = u0
    Niter = zeros(N)
    
    A = [1/4 1/4-1/6*sqrt(3); 1/4+1/6*sqrt(3) 1/4]
    b = [.5 .5]
    for i = 1:N
        k, Niter[i] = fixpoint([0.01,0.01], u[:,i], A, eps)
        u[:,i+1] = u[:,i] + h*b'.*k
    end
    return u,Niter
end

##############################
T = 60.0
dt = 0.01
N = round(Int64,T/dt)
t = collect(0:N)*dt
eps=0.1
w = pi
u0 = [2.0,0.0]

# GL 2-step fixed point
ugl2fixpoint,Niter = gl2fixpoint(N,T,u0,eps,500)

figure(1,figsize=(5,5))
plot(t,vec(ugl2fixpoint[1,:]),"r-")
xlabel(L"t")
ylabel(L"u(t)")

figure(2,figsize=(5,5))
plot(t[2:N+1],Niter,"b-")
xlabel(L"t")
ylabel(L"N_{iter}")
```
